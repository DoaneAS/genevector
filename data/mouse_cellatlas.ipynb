```python
from compass.data import Context, CompassDataset
from compass.model import Compass
import scanpy as sc
import warnings
warnings.filterwarnings('ignore')
```

```python
adata = sc.read("hvg_scaled_dataset2_cellatlas.h5ad")
adata = adata.raw.to_adata()
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var.highly_variable]
```

```python
context = Context.build(adata)
dataset = CompassDataset(context)
```

```python
context.frequency_histogram()
```

```python
context.metadata
```

```python
cmps = Compass(dataset,output_file="mouse_cellatlas.vec", batch_size=200)
```

```python
cmps.train(iterations=30)
```

```python
from compass.embedding import GeneEmbedding, CellEmbedding
embed = GeneEmbedding("mouse_cellatlas.vec", context)
```

```python
embed.compute_similarities("CD8A")
```

```python
embed.compute_similarities("C1QC")
```

```python
gene_clusters = embed.cluster()
```

```python
embed.plot(gene_clusters,labels=["C1QC","C1QA","TYROBP"])
```

```python
cembed = CellEmbedding(context, embed, adata)
```

```python
tsne_embedding = cembed.plot(column="cell_type")
```

```python
tsne_embedding = cembed.plot(column="batchlb",pcs=tsne_embedding)
```

```python
clusters = cembed.cluster(k=6)
```

```python
tsne_embedding = cembed.plot(pcs=tsne_embedding)
```

```python
cembed.batch_correct(column="batchlb",clusters=clusters)
```
